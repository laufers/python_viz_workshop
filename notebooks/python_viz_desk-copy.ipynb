## Line plots 

``` 
plot(x,y)
```
x: x coord
y: y coord

````python
# Create the data to be ploted

x = np.linspace(0,20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)
````

````python
# show x 

x
````

````python
# make the plot

plt.figure()

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
````

### Commonly used parameters:

    color: Set the color of the line.

    linestyle: Set the line style, e.g., solid, dashed, or none.

    linewidth: Set the line thickness.

    marker: Set the marker style, e.g., circles, triangles, or none.

    markersize: Set the marker size.

    label: Set the label for the line that will show up in the legend.



````python
plt.figure()

plt.plot(x, y1,
         color = 'black',
         linestyle = '-',
         linewidth =4,
         marker = 's', 
         markersize = 6,
         label = 'y1')

plt.plot(x, y2,
         color = 'green',
         linestyle = '-',
         linewidth =2,
         marker = 'd', 
         markersize = 6,
         label = 'y2')

plt.legend(loc = 'lower right')

plt.show()
````

## Vertical bar charts

bar(left, height) 


````python
# Create the dataset to be loted
years = np.arange(2012, 2015)
values = [2,5,9]
````

````python
plt.figure()

plt.bar(years, values)

plt.show()
````

### Commonly used parameters:

    color: Set the color of the bars.

    edgecolor: Set the color of the lines on the edges of the bars.

    width: Set the width of the bars.

    align: Set the alignment of the bars, e.g., center them on the x coordinate(s).

    label: Set the label for the bar that will show up in the legend.



````python
years = np.arange(2012, 2015)

category_1 = [2,5,9]
category_2 = [7, 6, 3]
````

````python
plt.figure()

plt.bar(years -.2,category_1,
       color = 'orange',
       edgecolor = 'none',
       width = 0.4,
       align = 'center', 
       label = 'y1')

plt.bar(years +.2,category_2,
       color = 'blue',
       edgecolor = 'none',
       width = 0.4,
       align = 'center', 
       label = 'y2')

plt.xticks(years, [str(year) for year in years])

plt.legend()

plt.show()
````

## horizontal bar charts

barh(bottom, width) 

````python
categories = ["A","B","C","D","E"]
values = [7,12,4,2,9]
````

````python
# Y axis ordered from A-E, top to bottom

plt.figure()

plt.barh(np.arange(len(categories)), values)

plt.yticks(np.arange(len(categories)), 
           ['Category {}'.format(x) for x in categories])

plt.show()
````

````python
# Y axis ordered from E-A, top to bottom

plt.figure()

plt.barh(-np.arange(len(categories)), values,
        color = 'yellow',
        edgecolor = 'purple',
        height = 0.6,
        align = 'center')

plt.yticks(-np.arange(len(categories)), ['Category {}'.format(x) for x in categories])

plt.show()
````

````python
# Sort according to size of values

# load pandas to make use of sort function
import pandas as pd

# Create dataframe
category_values = pd.DataFrame({'categories':categories,
                              'values':values})
````

````python
category_values
````

````python
category_values.sort_values(by='values', ascending = False, inplace = True)

categories = category_values['categories'].values
values = category_values['values'].values
````

````python
plt.figure()

plt.barh(-np.arange(len(categories)), values,
        color = 'yellow',
        edgecolor = 'purple',
        height = 0.6,
        align = 'center')

plt.yticks(-np.arange(len(categories)), ['Category {}'.format(x) for x in categories])

plt.show()
````

## Pie Charts

pie(sizes)  

````python
counts = [17,16, 9, 4, 1]

plt.figure()

plt.pie(counts)

plt.show()
````

````python
# change size of plotting area from default

plt.figure(figsize=(14,14))

plt.pie(counts)

plt.show()
````

### Commonly used parameters:

    colors: Set the colors of the wedges.

    labels: Set the labels of the wedges.

    startangle: Set the angle that the wedges start at.

    autopct: Set the percentage display format of the wedges.



````python
counts = [17,14]

plt.figure(figsize=(8,8))

plt.pie(counts, 
        colors = ['blue', 'orange'],
        labels = ['Category A', 'Category B'], 
                 startangle =90,
                 autopct='%1.1f%%')

plt.savefig('plots/pie_chart.pdf', dpi=150)

plt.show()
````

## Scater plots

matplotlib function: scatter(x, y)

    x, y: The values for the two variables.

````python
x = range(20)
y = np.arange(50,70) + (np.random.random(20) * 10)
````

````python
plt.figure()
plt.scatter(x,y)

plt.show()
````

### Commonly used parameters:

    c: Set the color of the markers.

    s: Set the size of the markers.

    marker: Set the marker style, e.g., circles, triangles, or squares.

    edgecolor: Set the color of the lines on the edges of the markers.



````python
plt.figure()

plt.scatter(x, y,
           c = 'red',
           s = 40,
           marker = 's',
           edgecolor = 'none')

plt.show()
````

## Histograms

Purpose: Showing the spread of a data column.

matplotlib function: hist(x)

    x: List of values to display the distribution of.


````python
# Create data

column_data = np.random.normal(42,3,1000)
````

````python
plt.figure()

plt.hist(column_data)

plt.show()
````

### Commonly used parameters:

    color: Set the color of the bars in the histogram.

    bins: Set the number of bins to display in the histogram, or specify specific bins.

    density: Indicate whether the bins should display fractions instead of counts.

````python
plt.figure()

plt.hist(column_data,
        color = 'cyan',
        bins = 50,
        density=True)

plt.show()
````

## Box plots

Purpose: Summarize the distribution of a data column.

matplotlib function: boxplot(x)

    x: List of values (or list of lists of values) to display the distribution of.


````python
# Create dataset

two_columns_data = [np.random.normal(57,3,1000),
                   np.random.normal(42,7,1000)]
````

````python
plt.figure()
plt.boxplot(two_columns_data)

plt.show()
````

### Commonly used parameters:

    labels: Set the labels for the box plots.

    notch: Indicate whether notches should be created around the median.

    bootstrap: Indicate whether the notches around the median should be bootstrapped.



````python
plt.figure() 
plt.boxplot(two_columns_data,
           labels=['A','B'],
           notch=True,
           bootstrap=10000)

plt.show()
````

## Subplots

Purpose: Allows you to place multiple charts in a figure.

matplotlib function: subplot(nrows, ncols, plot_number)

    nrows: The number of rows in the figure.

    ncols: The number of columns in the figure.

    plot_number: The placement of the chart (starts at 1).



````python
plt.figure()

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])
    
plt.tight_layout()

plt.show()
````

````python
# Create dataset

dist1 = np.random.normal(42,7,1000)
dist2 = np.random.normal(59,3,1000)
````

````python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(dist1)
plt.title('dist1')

plt.subplot(1,2,2)
plt.scatter(dist1, dist2)
plt.xlabel('dist1')
plt.ylabel('dist2')

plt.tight_layout()

plt.show()
````

````python
# Under plot area

years = np.arange(2010,2016)

plt.figure(figsize=(10,10))

for category_num in range(1,17):
    y_vals = np.arange(10,16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals, label = category_num)

plt.ylim(10,19)
plt.xlim(2010,2015)
plt.xticks(years, [str(year) for year in years])

plt.title('Legend under plot', fontsize= 18)
plt.ylabel('Counts', fontsize= 14)
plt.xlabel('Years', fontsize= 14)

plt.legend(bbox_to_anchor=(.5,-.1), loc = 'upper center', ncol = 8)

plt.savefig('plots/outside_legend01.png', dpi=150, bbox_inches='tight')
````

## Styles

````python
# Create three datasets

x1_values = [2012,2013,2014,2015]
y1_values = [4.3,2.5,1.8,4.5]

x2_values = [2012,2013,2014,2015]
y2_values = [2.4,4.4,1.8,2.8]

x3_values = [2012,2013,2014,2015]
y3_values = [2,2,3,5]
            
````

````python
plt.figure(figsize=(10,8))

plt.plot(x1_values,y1_values,
        label='Python',
        lw = 3,
        color = 'red')
plt.plot(x2_values,y2_values,
        label='JavaScript',
        lw = 3)
plt.plot(x3_values,y3_values,
        label = 'R',
        lw = 3)

plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
plt.yticks([1,2,3,4,5])

plt.xlim(2012,2015)
plt.ylim(0,6)

plt.xlabel('Year', fontsize=18)
plt.ylabel('Web Searches', fontsize = 18)

plt.legend(loc='upper center', ncol = 3, fontsize = 18)
plt.grid(True)

# plt.savefig('plots/web_search_01.png', dpi =150)

plt.show()
````

````python
plt.style.available
````

````python
plt.figure(figsize=(10,8))

with plt.style.context('ggplot'): 

    plt.plot(x1_values,y1_values,
            label='Python',
            lw = 3,
            color = 'red')
    plt.plot(x2_values,y2_values,
            label='JavaScript',
            lw = 3)
    plt.plot(x3_values,y3_values,
            label = 'R',
            lw = 3)

    plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
    plt.yticks([1,2,3,4,5])

    plt.xlim(2012,2015)
    plt.ylim(0,6)

    plt.xlabel('Year', fontsize=18)
    plt.ylabel('Web Searches', fontsize = 18)

    plt.legend(loc='upper center', ncol = 3, fontsize = 18)
    plt.grid(True)

    plt.savefig('plots/web_search_05.png', dpi =150)
    
    plt.show()
````

````python

# create a list of matplotlib styles
style_list = plt.style.available

plt.figure(figsize=(14,20))

# now loop over the list of styles and use subplot to arange output

counter = 1 # initialize counter

for style in style_list:

    with plt.style.context(style): 
 
        plt.subplot(8,4,counter)
        plt.plot(x1_values,y1_values,
                label='Python',
                lw = 2,
                color = 'red')
        plt.plot(x2_values,y2_values,
                label='JavaScript',
                lw = 2)
        plt.plot(x3_values,y3_values,
                label = 'R',
                lw = 2)

        plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
        plt.yticks([1,2,3,4,5])

        plt.xlim(2012,2015)
        plt.ylim(0,6)

        plt.xlabel('Year')
        plt.ylabel('Web Searches')
        plt.title(style)

        plt.legend(loc='upper center', ncol = 3, prop={'size': 6})
        plt.grid(True)
        
        counter += 1

plt.tight_layout()
plt.savefig('plots/styles.png', dpi =150)

plt.show()
````
