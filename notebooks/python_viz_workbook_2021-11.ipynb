# Legend Outside of plot area

````python
### Plot legend under plot area
````

````python
# Under plot area

years = np.arange(2010,2016)

plt.figure(figsize=(10,10))

for category_num in range(1,17):
    y_vals = np.arange(10,16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals, label = category_num)

plt.ylim(10,19)
plt.xlim(2010,2015)
plt.xticks(years, [str(year) for year in years])

plt.title('Legend under plot', fontsize= 18)
plt.ylabel('Counts', fontsize= 14)
plt.xlabel('Years', fontsize= 14)

plt.legend(bbox_to_anchor=(.5,-.1), loc = 'upper center', ncol = 8)

plt.savefig('plots/outside_legend01.png', dpi=150, bbox_inches='tight')
````

## Styles

````python
# Create three datasets

x1_values = [2012,2013,2014,2015]
y1_values = [4.3,2.5,1.8,4.5]

x2_values = [2012,2013,2014,2015]
y2_values = [2.4,4.4,1.8,2.8]

x3_values = [2012,2013,2014,2015]
y3_values = [2,2,3,5]
            
````

````python
plt.figure(figsize=(10,8))

plt.plot(x1_values,y1_values,
        label='Python',
        lw = 3,
        color = 'red')
plt.plot(x2_values,y2_values,
        label='JavaScript',
        lw = 3)
plt.plot(x3_values,y3_values,
        label = 'R',
        lw = 3)

plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
plt.yticks([1,2,3,4,5])

plt.xlim(2012,2015)
plt.ylim(0,6)

plt.xlabel('Year', fontsize=18)
plt.ylabel('Web Searches', fontsize = 18)

plt.legend(loc='upper center', ncol = 3, fontsize = 18)
plt.grid(True)

# plt.savefig('plots/web_search_01.png', dpi =150)

plt.show()
````

````python
plt.figure(figsize=(10,8))

with plt.style.context('ggplot'): 

    plt.plot(x1_values,y1_values,
            label='Python',
            lw = 3,
            color = 'red')
    plt.plot(x2_values,y2_values,
            label='JavaScript',
            lw = 3)
    plt.plot(x3_values,y3_values,
            label = 'R',
            lw = 3)

    plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
    plt.yticks([1,2,3,4,5])

    plt.xlim(2012,2015)
    plt.ylim(0,6)

    plt.xlabel('Year', fontsize=18)
    plt.ylabel('Web Searches', fontsize = 18)

    plt.legend(loc='upper center', ncol = 3, fontsize = 18)
    plt.grid(True)

    plt.savefig('plots/web_search_05.png', dpi =150)
    
    plt.show()
````

````python

# create a list of matplotlib styles
style_list = plt.style.available

plt.figure(figsize=(14,20))

# now loop over the list of styles and use subplot to arange output

counter = 1 # initialize counter

for style in style_list:

    with plt.style.context(style): 
 
        plt.subplot(8,4,counter)
        plt.plot(x1_values,y1_values,
                label='Python',
                lw = 2,
                color = 'red')
        plt.plot(x2_values,y2_values,
                label='JavaScript',
                lw = 2)
        plt.plot(x3_values,y3_values,
                label = 'R',
                lw = 2)

        plt.xticks([2012,2013,2014,2015],['2012', '2013', '2014', '2015'] )
        plt.yticks([1,2,3,4,5])

        plt.xlim(2012,2015)
        plt.ylim(0,6)

        plt.xlabel('Year')
        plt.ylabel('Web Searches')
        plt.title(style)

        plt.legend(loc='upper center', ncol = 3, prop={'size': 6})
        plt.grid(True)
        
        counter += 1

plt.tight_layout()
plt.savefig('plots/styles.png', dpi =150)

plt.show()
````
